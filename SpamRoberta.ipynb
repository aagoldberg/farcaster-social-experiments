```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("mshenoda/roberta-spam")
model = AutoModelForSequenceClassification.from_pretrained("mshenoda/roberta-spam")
```

```python
import torch
```

```python
# Sample text to test the model
texts = [
    "Congratulations! You've won a $1,000 gift card. Click here to claim your prize.",
    "Hey, are we still on for lunch tomorrow?",
    "Urgent! Your account has been compromised. Please login immediately to secure your account.",
    "Great tips! I'll definitely take advantage of the low gas fees to renew my ENS. Happy trading!",
    "You are selected for a free trial subscription to our premium service. No credit card required!",
    "Revolutionary solution for cross-chain payments. Easy, seamless transactions with any token.",
    "Excited for the design channel! Let's see what you guys come up with! 🔥"
]

# Tokenize the texts
inputs = tokenizer(texts, return_tensors="pt", truncation=True, padding=True)

# Perform inference
with torch.no_grad():
    outputs = model(**inputs)
    predictions = torch.argmax(outputs.logits, dim=-1)

# Convert predictions to readable format
labels = ["Not Spam", "Spam"]
results = [labels[pred] for pred in predictions]

for text, result in zip(texts, results):
    print(f"Text: {text}\nPrediction: {result}\n")
```
