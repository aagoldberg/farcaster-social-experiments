```python
df = pd.read_csv(f'/content/drive/My Drive/repliesWithHidden_final.csv')  # Replace with the path to your CSV
df['text'] = df['text'].astype(str)
```

```python
df.head()
```

```python
model_name = "roberta-base-openai-detector"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# texts = [
#     "The sun rises in the east and sets in the west.",
#     "Once upon a time, in a land far, far away, there was a magical kingdom.",
#     "Our company is excited to announce a new AI tool that will revolutionize content creation.",
#     "Hello, how can I assist you today?"
# ]
```

```python
inputs = tokenizer(df['text'].tolist(), padding=True, truncation=True, return_tensors="pt")
```

```python
# Tokenize the input texts
# inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
```

```python
# Perform inference
with torch.no_grad():
    outputs = model(**inputs)
    predictions = torch.softmax(outputs.logits, dim=-1)

# Extract the probabilities of being "real" or "fake"
df['real_prob'] = predictions[:, 0].tolist()
df['fake_prob'] = predictions[:, 1].tolist()

```

```python
df.to_csv(f'/content/drive/My Drive/repliesWithRealFakeProb.csv')
```
